```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Normalize the email column in both
leads_df['Registered Email'] = leads_df['Registered Email'].str.lower().str.strip()
apps_df['Registered Email'] = apps_df['Registered Email'].str.lower().str.strip()

# Extract all unique years from leads
years = sorted(leads_df['Lead Year'].dropna().unique())

# Prepare summary list
year_stats = []

for year in years:
    leads_year = leads_df[leads_df['Lead Year'] == year]
    leads_emails = leads_year['Registered Email'].dropna().unique()

    # Match against all applicants
    matched_emails = np.intersect1d(leads_emails, apps_df['Registered Email'].dropna().unique())

    total_leads = len(leads_emails)
    matched = len(matched_emails)
    not_matched = total_leads - matched

    year_stats.append({
        'Year': year,
        'Total Leads': total_leads,
        'Converted': matched,
        'Not Converted': not_matched,
        'Conversion Rate (%)': round((matched / total_leads) * 100, 2) if total_leads > 0 else 0
    })

# Create summary DataFrame
year_stats_df = pd.DataFrame(year_stats)
print(year_stats_df)
```

```python
# Bar chart
plt.figure(figsize=(10, 6))
x = np.arange(len(year_stats_df))
bar_width = 0.4

plt.bar(x - bar_width/2, year_stats_df['Converted'], width=bar_width, label='Converted', color='green')
plt.bar(x + bar_width/2, year_stats_df['Not Converted'], width=bar_width, label='Not Converted', color='red')

plt.xticks(x, year_stats_df['Year'])
plt.ylabel('Number of Leads')
plt.title('Lead Conversion by Year')
plt.legend()
plt.grid(True, axis='y', linestyle='--', alpha=0.5)
plt.tight_layout()
plt.show()

# Line chart: conversion rate
plt.figure(figsize=(10, 4))
plt.plot(year_stats_df['Year'], year_stats_df['Conversion Rate (%)'], marker='o', color='blue')
plt.title('Conversion Rate (%) by Year')
plt.xlabel('Year')
plt.ylabel('Conversion Rate (%)')
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Step 1: Clean emails for matching
leads_df['Registered Email'] = leads_df['Registered Email'].str.lower().str.strip()
apps_df['Registered Email'] = apps_df['Registered Email'].str.lower().str.strip()

# Step 2: Mark conversions
converted_emails = set(apps_df['Registered Email'].dropna())
leads_df['Converted'] = leads_df['Registered Email'].isin(converted_emails)

# Step 3: Fix 'Lead Year' column
leads_df['Lead Year'] = pd.to_numeric(leads_df['Lead Year'], errors='coerce')
leads_df = leads_df.dropna(subset=['Lead Year'])  # Drop rows with missing year
leads_df['Lead Year'] = leads_df['Lead Year'].astype(int)

# Step 4: Fill missing lead owners (optional)
leads_df['Lead Owner'] = leads_df['Lead Owner'].fillna('Unknown')
```

```python
# Group and calculate conversion metrics
conversion_summary = leads_df.groupby(['Lead Year', 'Publisher Name']).agg(
    Total_Leads=('Registered Email', 'count'),
    Converted_Leads=('Converted', 'sum')
).reset_index()

# Calculate conversion rate
conversion_summary['Conversion_Rate (%)'] = (
    conversion_summary['Converted_Leads'] / conversion_summary['Total_Leads'] * 100
).round(2)

# Sort for easy viewing
conversion_summary = conversion_summary.sort_values(['Lead Year', 'Conversion_Rate (%)'], ascending=[True, False])

# View top 10 rows
print(conversion_summary.head(10))
```

```python
plt.figure(figsize=(14, 8))
sns.set(style="whitegrid")

# Filter: Only show top 10 lead owners by volume across years
top_owners = (
    conversion_summary.groupby('Publisher Name')['Total_Leads'].sum()
    .sort_values(ascending=False)
    .head(10).index
)
plot_df = conversion_summary[conversion_summary['Publisher Name'].isin(top_owners)]

# Plot
sns.barplot(
    data=plot_df,
    x='Publisher Name',
    y='Conversion_Rate (%)',
    hue='Lead Year',
    palette='Set2'
)

plt.title('Conversion Rate by Publisher Name (All Years)', fontsize=16)
plt.ylabel('Conversion Rate (%)')
plt.xlabel('Publisher Name')
plt.xticks(rotation=55)
plt.legend(title='Year')
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import numpy as np

# Normalize emails
leads_df['Registered Email'] = leads_df['Registered Email'].str.lower().str.strip()
apps_df['Registered Email'] = apps_df['Registered Email'].str.lower().str.strip()

# Drop rows with missing state or email
leads_df = leads_df.dropna(subset=['State', 'Registered Email', 'Lead Year'])

# Ensure Lead Year is int
leads_df['Lead Year'] = pd.to_numeric(leads_df['Lead Year'], errors='coerce').astype(int)

# Unique list of years and states
years = sorted(leads_df['Lead Year'].unique())
apps_emails = set(apps_df['Registered Email'].dropna())

# Store results
results = []

for year in years:
    leads_year = leads_df[leads_df['Lead Year'] == year]
    for state in leads_year['State'].unique():
        leads_state = leads_year[leads_year['State'] == state]
        state_emails = set(leads_state['Registered Email'])

        matched = len(state_emails & apps_emails)
        total = len(state_emails)
        not_matched = total - matched
        rate = round((matched / total) * 100, 2) if total > 0 else 0

        results.append({
            'Year': year,
            'State': state,
            'Total Leads': total,
            'Converted': matched,
            'Not Converted': not_matched,
            'Conversion Rate (%)': rate
        })

# Create final DataFrame
state_year_df = pd.DataFrame(results)
print(state_year_df.head())
```

```python
# Step 1: Normalize the 'Email' column in both datasets
leads_df['Registered Email'] = leads_df['Registered Email'].str.lower().str.strip()
apps_df['Registered Email'] = apps_df['Registered Email'].str.lower().str.strip()

# Step 2: Create a list of emails from the application data (converted leads)
apps_emails = set(apps_df['Registered Email'].dropna())

# Step 3: Group the leads data by State and Year, and count conversions
conversion_results = []

for year in sorted(leads_df['Lead Year'].unique()):
    leads_year = leads_df[leads_df['Lead Year'] == year]
    
    for state in leads_year['State'].unique():
        leads_state = leads_year[leads_year['State'] == state]
        state_emails = set(leads_state['Registered Email'])

        # Count matched emails (conversions)
        matched = len(state_emails & apps_emails)
        
        conversion_results.append({
            'Year': year,
            'State': state,
            'Conversions': matched
        })

# Create the DataFrame with conversion counts
conversion_df = pd.DataFrame(conversion_results)

# Display the result
print(conversion_df.head())
```
